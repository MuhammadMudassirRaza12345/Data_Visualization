```python
#Categorical =>bar plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
titani=sns.load_dataset("titanic")
sns.catplot(x="sex",y="survived",hue="class",kind="bar",data=titani)
plt.show()
```

```python
#Categorical =>countplots
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
titanic=sns.load_dataset("titanic")
p1=sns.countplot(x="sex", hue="class", data=titanic)
p1.settitle("Plot for counting")
plt.show()
```

```python
#Contionous  =>scatter plot
 
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
titanic=sns.load_dataset("titanic")
g=sns.FacetGrid(row="sex", hue="alone", data=titanic)
g=(g.map(plt.scatter,"age","fare" ).add_legend())
g.settitle("Plot for counting")
plt.show()
```

```python
#data visualization
#step1:importing libraries
import seaborn as sns
import matplotlib.pyplot as plt

# step#3 set a theme
sns.set_theme(style="ticks",color_codes=True)

#step3:loading dataset
kashti=sns.load_dataset("titanic")
print(kashti)
# print(kashti.describe())    

# #step4:plot the basic graph one variables (count plot)
p=sns.countplot(x="sex" ,data=kashti)
#count plot categroical data

plt.show()

# #step5:plot the basic graph with two vqriables (count plot)
#hue mean colours
p=sns.countplot(x="sex" ,data=kashti,hue="class")
#count plot categroical data
plt.show()

#step6:plot the basic graph with two vqriables (count plot) with titles
p=sns.countplot(x="sex" ,data=kashti,hue="class")
#count plot categroical data
p.set_title("count plot fot titanic")
plt.show()
```
